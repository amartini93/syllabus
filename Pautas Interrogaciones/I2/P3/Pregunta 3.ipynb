### b) Explique para qué sirven los métodos ***__enter__*** y ***__exit__*** en una clase cualquiera de Python

###Respuesta
Un **context manager** es un objeto que define el contexto de ejecución que se establece cuando se ejecuta un **with**. Este se preocupa de ejecutar los métodos **\_\_enter\_\_** y **\_\_exit\_\_** cuando corresponda. 

**\_\_enter\_\_**: define lo que el context manager debe hacer al inicio del bloque creado por **with**. Lo que retorna está restringido por el target del **with**. 

**\_\_exit\_\_**: define lo que el context manager debe hacer después de que el bloque creado por **with** haya terminado de ejecutarse o que haya lanzado un error. 

###Puntaje
- Si solo puso qué son => 0.5 punto c/u
- Si puso cómo funcionan => 1.5 punto c/u
- Si mencionan al context manager 1 punto

```python
class Dog():
    def __enter__(self):
        print("Inicio")
        return self #Si no retorna nada, entonces doggy será None
    
    def __exit__(self, type, value, tb):
        print("Fin")
        
    def hola(self):
        print("guau!")

class Cat():
    def hola(self):
        print("miau!")

d = Dog()
c = Cat()
with d as doggy:
    doggy.hola()

with c as kitty: #Se lanza un error porque Cat no tiene implementado los métodos enter y exit
    kitty.hola()
```
